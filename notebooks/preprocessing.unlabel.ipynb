```python
import pandas as pd
import re
```

```python
df = pd.read_csv('./samples/pet.unlabel.csv')
df.head(10)
```

```python
def reduce_texts(x):
    x = str(x)
    x = x.strip(r' \t\n')
    x = re.sub(r'\s+', ' ', x)
    x = re.sub(r'[0-9]|[$-_@.&+!*?;:~\(\)\[\]<>/\'"\\]+', '', x)
    return '' if x.isspace() or len(x) < 8 else x

with open('corpus.txt', 'w') as out:
    for index, row in df.iterrows():
        if type(row['SE']) == str and not row['SE'].isspace():
            print(reduce_texts(row['SE']), file=out)
```
