```
class PetDataset(Dataset):
  def __init__(self, df):
    self.df = df
    
  def __len__(self):
    return len(self.df)

  def __getitem__(self, idx):
    text = self.df.iloc[idx, 9]
    label = self.df.iloc[idx, 10]
    return text, label
```

```
train_dataset = PetDataset(train_df)
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=False, num_workers=1)
```

```
device = torch.device("cuda")
tokenizer = RobertaTokenizer.from_pretrained("./pretrained", do_lower_case=False)
```

```
iterator = iter(train_loader)
for _ in range(100):
  text, label = next(iterator)
  output = tokenizer.encode(text[0])
  print(len(output), tokenizer.decode(output))
```
