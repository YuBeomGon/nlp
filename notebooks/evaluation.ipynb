```python
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification

device = torch.device("cpu")
tokenizer = RobertaTokenizer.from_pretrained("./pretrained", do_lower_case=False)
model = RobertaForSequenceClassification.from_pretrained("./pretrained", num_labels=26)
model.to(device)
```

```python
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

class PetDataset(Dataset):
    def __init__(self, df):
        self.df = df

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        text = self.df.iloc[idx, 9]
        label = self.df.iloc[idx, 10]
        return text, label
    
def evaluate(df):
    total_loss = 0
    total_len = 0
    total_correct = 0
    
    dataset = PetDataset(df)
    loader = DataLoader(dataset, batch_size=1, shuffle=False, num_workers=2)
    
    model.eval()
    
    for text, label in loader:
        encoded_list = [tokenizer.encode(t, add_special_token=True) for t in text]
        padded_list = [e[:512] + [0] * (512 - len(e[:512])) for e in encoded_list]
        sample = torch.tensor(padded_list)
        sample, label = sample.to(device), label.to(device)
        labels = torch.tensor(label)
        outputs = model(sample, labels=labels)
        _, logits = outputs
        
        pred = torch.argmax(F.softmax(logits), dim=1)
        correct = pred.eq(labels)
        total_correct += correct.sum().item()
        total_len += len(labels)
        
    return total_correct / total_len
```

```python
score = evaluate(test_df)
print("ALL = {}".format(score))
```

```python
eval_df = pd.DataFrame(columns=('code', 'score', 'count'))
```

```python
for code in test_df.groupby('주진단코드').groups.keys():
    score = evaluate(test_df[test_df['주진단코드'] == code])
    count = count_df['SE'][str(code)]
    eval_df = eval_df.append({'code': code, 'score': score, 'count': count}, ignore_index=True)
    print("{} = {} ({})".format(code, score, count))
```
