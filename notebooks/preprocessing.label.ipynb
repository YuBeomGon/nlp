```python
import pandas as pd
import re
```

```python
df = pd.read_csv('./samples/pet.label.csv')
df.head(10)
```

```python
df.groupby('병원').count()
```

```python
df.groupby('주진단코드').count()
```

```python
df = df.groupby('주진단코드').filter(lambda g: len(g) > 10)
df.groupby('주진단코드').count()
```

```python
def reduce_codes(x):
    x = str(x)
    x = x.strip(r' \t\n')
    x = re.sub(r'\s+', ' ', x)
    x = re.sub(r'중성화수술','neutral', x)
    x = re.sub(r'건강검진','test', x)
    x = re.sub(r'XX|X|x|-|이상없음','normal', x)
    x = re.sub(r'[.]0','', x)
    return '' if x.isspace() or len(x) < 3 else x
df['주진단코드'] = df['주진단코드'].apply(lambda x: reduce_codes(x))
df = df[(df['주진단코드'] != '') & (df['주진단코드'] != 'nan')]
```

```python
df[df['주진단코드'] == '2047'] = '2043'
df[df['주진단코드'] == '2048'] = '2043'
```

```python
def reduce_texts(x):
    x = str(x)
    x = x.strip(r' \t\n')
    x = re.sub(r'\s+', ' ', x)
    x = re.sub(r'[0-9]|[$-_@.&+!*?;:~\(\)\[\]<>/\'"\\]+', '', x)
    return '' if x.isspace() or len(x) < 8 else x
df['SE'] = df['SE'].apply(lambda x: reduce_texts(x))
df = df[(df['SE'] != '') & (df['SE'] != 'nan')]
```

```python
df.groupby('병원').count()
```

```python
df.groupby('주진단코드').count()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(df['주진단코드'])
encoded = encoder.transform(df['주진단코드'])
df['ID'] = encoded
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.1, random_state=900)
```

```python
train_df.groupby('병원').count()
```

```python
train_df.groupby('ID').count()
```

```python
train_df.to_csv('train.csv', index=None)
test_df.to_csv('test.csv', index=None)
```
